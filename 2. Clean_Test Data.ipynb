```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("dataset.csv")
df
```

```python
col=list(df.columns)
len(col)
```

```python
s=[]
for i in range (1,len(col)):
    s.append(list(df[col[i]].unique()))
```

```python
p=set()
for i in range(0,17):
    p=p.union(set(s[i]))
p=list(p)
```

```python
d=np.array(df.groupby('Disease').size().index)
```

```python
len(p)
```

```python
len(d)
```

```python
data=pd.DataFrame(columns=p)
```

```python
data
```

```python
for k in range (0,len(df)):
    l=[0]*132
    v=list(df.iloc[k].values)
    for i in range (0,18):
        for j in range (0,132):
            if p[j]==v[i]:
                l[j]=1
    data.loc[k]=l
    
```

```python
data["Disease"]=df["Disease"]
```

```python
encode={}
for i,j in enumerate(d):
    encode[j]=i
encode
```

```python
for x in range(0,len(data)):
    data["Disease"].iloc[x]=encode[ data["Disease"].iloc[x]]
    
```
