```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("dataset.csv")
df
```

```python
col=list(df.columns)
col
```

```python
def unique_values(b,a=[],c=set()):
    global col
    a=[]
    for i in range(1,len(col)):
        a.append(list(b[col[i]].unique()))
    #print(a)
    c=set()
    for i in range(len(col)-1):
        c=c.union(set(a[i]))
    return list(c)
    
```

```python
unique_values(df[df['Disease']=='(vertigo) Paroymsal  Positional Vertigo'])
```

```python
disease=list(np.array(df.groupby('Disease').size().index))
disease[0]
```

```python
symptoms=unique_values(df)
symptoms
```

```python
main={}
for i,j in enumerate(disease):
    # print(i,j)
    main[i]=unique_values(df[df['Disease']==j])
main
```

```python
disease_dict={}
for i,j in enumerate(disease):
    disease_dict[i]=j
disease_dict
```

```python
data=pd.DataFrame(columns=symptoms)
data
```

```python
for k in range(len(disease)):
    l=[0]*132
    for i in range(len(main[k])):
        for j in range(len(symptoms)):
            if symptoms[j]==main[k][i]:
                l[j]=1
    data.loc[k]=l
```

```python
data
```

```python
data["Disease"]=data.index
```

```python
data.to_csv('train_jyoti.csv')
```
